Data--> Data Preprocessing --> Feature Extraction --> User Input --> Cosine Similarity Algorithm --> List of Movies 

```python
import numpy as np
import pandas  as  pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

Data Collection And Pre-Processing

Loading the Data from CSV file  to Pandas Frame

```python
movie_data = pd.read_csv('movies.csv')
```

```python
movie_data.head()
```

```python
movie_data.shape
```

Feature Extraction --> Selecting the relevant feature for selection

```python
features_selected=['genres','keywords','title','original_language','tagline','cast','director']
print(features_selected)
```

Replacing Null Values with Null String

```python
for feature in features_selected:
    movie_data[feature]=movie_data[feature].fillna('')
```

Combining all the selelcted feature

```python
combine_feature = movie_data['genres']+' '+movie_data['keywords']+' '+movie_data['title']+' '+movie_data['original_language']+' '+movie_data['tagline']+' '+movie_data['cast']+' '+movie_data['director']
print(combine_feature)
```

Converting Text Data to Feature Vectors

```python
vectorizer= TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combine_feature)
```

```python
print(feature_vectors)
```

Cosine Similarity

```python
similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

```python
print(similarity.shape)
```

```python
movie_name = 'iron man'
```

```python
list_movies = movie_data['title'].tolist()
print(list_movies)
```

Finding the best match given by user

```python
Close_match  = difflib.get_close_matches(movie_name, list_movies)
print(Close_match)
```

```python
close_matching = Close_match[0]
print(close_matching)
```

```python
index_of_movie  = movie_data[movie_data.title == close_matching]['index'].values[0]
print(index_of_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

sorting movies based on the similarity score 

```python
sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse= True)
print(sorted_similar_movies)
```

```python
print('Movies Suggested for You: \n')
i=1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movie_data[movie_data.index==index]['title'].values[0]
    if (i<30):
        print(i,'.', title_from_index)
        i+=1
```

Movies Recommendation System

```python
movie_name = 'avatar'
list_movies = movie_data['title'].tolist()
Close_match  = difflib.get_close_matches(movie_name, list_movies)
close_matching = Close_match[0]
index_of_movie  = movie_data[movie_data.title == close_matching]['index'].values[0]
similarity_score = list(enumerate(similarity[index_of_movie]))
sorted_similar_movies = sorted(similarity_score, key=lambda x:x[1], reverse= True)
print(sorted_similar_movies)
print('Movies Suggested for You: \n')
i=1
for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movie_data[movie_data.index==index]['title'].values[0]
    if (i<30):
        print(i,'.', title_from_index)
        i+=1
```
